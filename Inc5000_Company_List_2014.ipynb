```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


plt.figure(figsize=(8,5))
plt.hist(df["revenue_usd"], bins=30)
plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.show()
```

### Observation

- The revenue distribution is right-skewed.
- Most companies generate relatively lower revenue.
- Only a small number of companies generate extremely high revenue.

```python
plt.figure(figsize=(8,5))
plt.hist(df["growth_percent"], bins=30)
plt.title("Growth Distribution")
plt.xlabel("Growth (%)")
plt.ylabel("Frequency")
plt.show()
```

### Observation

- Most companies have moderate growth rates.
- A few companies exhibit exceptionally high growth.
- High-growth companies appear as outliers in the distribution.

```python
plt.figure(figsize=(14,6))

top_industries = df["industry"].value_counts().head(10)

top_industries.plot(kind="bar")

plt.title("Top 10 Industries")
plt.xlabel("Industry")
plt.ylabel("Number of Companies")

plt.xticks(rotation=45, ha='right')

plt.tight_layout()

plt.show()
```

### Observation

- IT Services has the highest number of companies.
- Advertising & Marketing and Software are also among the most represented industries.
- Technology-related industries dominate the dataset.

```python
top_states = df["state"].value_counts().head(10)

plt.figure(figsize=(10,6))
top_states.plot(kind="bar")
plt.title("Top 10 States")
plt.xlabel("State")
plt.ylabel("Number of Companies")
plt.xticks(rotation=45)
plt.show()
```

### Observation

- A small number of states contain the majority of companies.
- Business activity is concentrated in economically developed regions.

```python
plt.figure(figsize=(8,6))
plt.scatter(df["employee_count"], df["revenue_usd"])
plt.title("Revenue vs Employee Count")
plt.xlabel("Employee Count")
plt.ylabel("Revenue")
plt.show()
```

### Observation

- There is a positive relationship between employee count and revenue.
- Companies with more employees generally report higher revenue.
- Some companies generate high revenue despite having relatively few employees.

```python
plt.figure(figsize=(10,6))

df.groupby('industry')['revenue_usd'].mean().sort_values().tail(10).plot(kind='barh')

plt.title("Top 10 Industries by Average Revenue")
plt.xlabel("Average Revenue (USD)")
plt.ylabel("Industry")

plt.tight_layout()
plt.show()
```

### Observation

- Average revenue differs across industries.
- Some industries with fewer companies generate higher average revenue.
- Revenue performance is influenced by industry type.
